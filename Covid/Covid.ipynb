```python
import re
import requests
import pandas as pd

from bs4 import BeautifulSoup
```

```python
url="https://www.worldometers.info/coronavirus/#google_vignette"
```

```python
request_header= {'Content-Type':'text/css; charset=UTF-8',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',
     'Accept-Encoding': 'gzip, deflate, br'
}
```

```python
response=requests.get(url,headers=request_header)
```

```python
print(response.status_code)
```

```python
html_code=response.text
```

```python
soup=BeautifulSoup(html_code)
```

```python


table = soup.find('table', id="main_table_countries_today")
table
```

```python
headers = []
for i in table.find_all('th'):
    title = i.text
    headers.append(title)
```

```python
headers
```

```python

headers[13] = 'Tests/1M pop'
```

```python

data = pd.DataFrame(columns =headers)
```

```python
data
```

```python

for j in table.find_all('tr')[1:]:
    row_data = j.find_all('td')
    row = [i.text for i in row_data]
    length = len(data)
    data.loc[length] = row
```

```python

data.drop(data.index[0:7], inplace=True)
data.drop(data.index[222:229], inplace=True)
data.reset_index(inplace=True, drop=True)

```

```python

data.to_csv('covid_data.csv', index=False)

dataset = pd.read_csv('covid_data.csv')
```
