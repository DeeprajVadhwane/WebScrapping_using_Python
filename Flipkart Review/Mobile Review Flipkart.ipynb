# Extracting Data using find_all()

```python
%%time
import requests
from bs4 import BeautifulSoup
import numpy as np

reviewer_name = []
review_rating = []
review_title = []
review_text = []
review_place = []
review_date = []
up_votes = []
down_votes = []

pages = 1328

print("Progress:", end="")

for i in range(1, pages):
    url = "https://www.flipkart.com/apple-iphone-13-green-128-gb/product-reviews/itm18a55937b2607?pid=MOBGC9VGSU9DWGJZ&lid=LSTMOBGC9VGSU9DWGJZTOZYKQ&marketplace=FLIPKART"
    request_header = {
        'Content-Type': 'text/html; charset=UTF-8',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0',
        'Accept-Encoding': 'gzip, deflate, br'
    }

    response = requests.get(url, headers=request_header)
    html_code = response.text

    soup = BeautifulSoup(html_code)

    reviewer_names_data = soup.find_all('p', class_="_2sc7ZR _2V5EHH")
    for name in reviewer_names_data:
        reviewer_name.append(name.text)

    review_ratings_data = soup.find_all('div', class_="_3LWZlK _1BLPMq")
    for rating in review_ratings_data:
        if rating is None:
            review_rating.append(None)
        else:
            review_rating.append(rating.text)

    titles = soup.find_all('p', class_="_2-N8zT")
    for title in titles:
        review_title.append(title.text)

    review_texts_data = soup.find_all('div', class_="t-ZTKy")
    for text in review_texts_data:
        review_text.append(text.text)

    place_of_reviews_data = soup.find_all('p', class_="_2mcZGG")
    for place in place_of_reviews_data:
        review_place.append(place.text)
    
    review_data = soup.find_all('p', class_="_2sc7ZR")
    for idx in range(0, len(review_data), 2):
        
        # Extract date of review
        review_date.append(review_data[idx + 1].text)

    down_votes_data = soup.find_all('div', class_="_1LmwT9 pkR4jH")
    for down_vote in down_votes_data:
        down_votes.append(down_vote.text)

    for container in soup.find_all('div', class_='_27aTsS'):
        product = container.find('span', class_='_3c3Px5')
        if product is None:
            up_votes.append(None)
        else:
            up_votes.append(product.text)

    print(".", end="")

print()
```

```python

print("Length of reviewer_names:", len(reviewer_name))
print("Length of review_ratings:", len(review_rating))
print("Length of review_titles:", len(review_title))
print("Length of review_texts:", len(review_text))
print("Length of place_of_reviews:", len(review_place))
print("Length of date_of_reviews:", len(review_date))
print("Length of up_votes:", len(up_votes))
print("Length of down_votes:", len(down_votes))
```

```python
import pandas as pd

# Assuming you have the extracted data in the lists as shown in the previous code

# Create a dictionary with the data
data = {
    "Reviewer Name": reviewer_name,
    "Reviewer Rating": review_rating,
    "Review Title": review_title,
    "Review Text": review_text,
    "Place of Review": review_place,
    "Date of Review": review_date,
    "Up Votes": up_votes,
    "Down Votes": down_votes
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Save the DataFrame to a CSV file
df.to_csv("flipkart_reviews.csv", index=False)
```
